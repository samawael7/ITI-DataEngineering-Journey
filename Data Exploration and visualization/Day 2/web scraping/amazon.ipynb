```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import pandas as pd
```

```python
scraper = webdriver.Chrome()
scraper.get("https://www.amazon.eg/s?i=electronics&rh=n%3A21832883031%2Cp_123%3A46655&language=en&pf_rd_i=21832883031&pf_rd_m=A1ZVRGNO5AYLOV&pf_rd_p=deca76a5-c5fe-453b-ac5f-5b72e8f03038&pf_rd_r=T6TVJ323N65TPZG39GBH&pf_rd_s=merchandised-search-9&ref=wirelessbrands_EG_1")
```

```python
items=scraper.find_elements(By.XPATH,"//h2[@class='a-size-base-plus a-spacing-none a-color-base a-text-normal']/span")
prices = scraper.find_elements(By.XPATH, "//span[@class='a-price-whole']")
```

```python
for item in items:
    print(item.text)
for price in prices:
    print(price.text)
print(len(items))
print(len(prices))
```
